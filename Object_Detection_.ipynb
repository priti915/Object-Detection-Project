```python
# STEP 2: Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
import numpy as np
```

```python
# STEP 3: Upload image (you can skip this if using a fixed image)
uploaded = files.upload()
for fn in uploaded.keys():
    img_path = fn

# STEP 4: Load the image
image = cv2.imread(img_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# STEP 5: Load pre-trained face detector (Haar cascade)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# STEP 6: Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# STEP 7: Draw rectangle around faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# STEP 8: Show image with detected face
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
